```python
import pandas as pd

df = pd.read_excel('omloopplanning.xlsx') #excel dataframe inladen met pandas
```

```python
df[['starttijd datum','eindtijd_datum']] = df[['starttijd datum','eindtijd datum']].apply(pd.to_datetime) #data omzetten van object naar datetime

df['startlocatie_index'] = df['startlocatie'].copy() #nieuw kolommen maken om te vervangen met 0, 1 en 2
df['eindlocatie_index'] = df['eindlocatie'].copy() 

df['eindlocatie_index'] = df['eindlocatie'].replace(['ehvgar','ehvbst','ehvapt'],[0,1,2]) #vervangen van garage voor 0, busstation voor 1 en airport voor 2
df['startlocate_index'] = df['startlocatie'].replace(['ehvgar','ehvbst','ehvapt'],[0,1,2])
```

```python
import matplotlib.pyplot as plt

plt.plot(df['starttijd datum'][df['omloop nummer'] == 1],df['startlocate_index'][df['omloop nummer'] == 1]) #plotten van waar bus van omloop 1 is op welke tijd
plt.xlabel('tijd')
plt.ylabel('locatie index')
```
